```python
import boto3
import io
import pandas as pd

# reference: https://dluo.me/s3databoto3
resource = 's3'
bucket_name = 'safesense-audio'
dir_csv = 'foreground_dog_bark' # change based on desired audio to retrieve

client = boto3.client(resource) #low-level functional API
resource = boto3.resource(resource) #high-level object-oriented API
bucket = resource.Bucket(bucket_name) #S3 bucket
```

```python
# retrieve type csv
response = client.get_object(Bucket=bucket_name, Key=dir_csv+ '/' + dir_csv + '.csv')
file = io.BytesIO(response.get('Body').read())
pd.read_csv(file)

# to upload files
# bucket.upload_file('file',Key='path/to/my/file')

list(bucket.objects.filter(Prefix=''))
```
